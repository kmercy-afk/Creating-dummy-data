```python
import numpy as np
import matplotlib.pyplot as plt

# Mean vector and covariance matrix
mean = [-3, 0]
covariance_matrix = [
    [1.0, 0.8],
    [0.8, 1.0]
]

# Generate 500 random points with the specified distribution
data = np.random.multivariate_normal(mean, covariance_matrix, 500)
x, y = data[:, 0], data[:, 1]

# Determine the range for x-axis alignment
x_min, x_max = min(x.min(), y.min()), max(x.max(), y.max())

# Create two subplots for histograms
fig, axs = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Histogram for x dimension
axs[0].hist(x, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
axs[0].set_title("Histogram of X Dimension", fontsize=14)
axs[0].set_xlabel("X Values", fontsize=12)
axs[0].set_ylabel("Frequency", fontsize=12)
axs[0].grid(alpha=0.3)
axs[0].set_xlim(x_min, x_max)

# Histogram for y dimension
axs[1].hist(y, bins=30, color='lightcoral', edgecolor='black', alpha=0.7)
axs[1].set_title("Histogram of Y Dimension", fontsize=14)
axs[1].set_xlabel("Y Values", fontsize=12)
axs[1].grid(alpha=0.3)
axs[1].set_xlim(x_min, x_max)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Question 1: Data (mean = [-3, 0])
mean_q1 = [-3, 0]
covariance_matrix_q1 = [
    [1.0, 0.8],
    [0.8, 1.0]
]
data_q1 = np.random.multivariate_normal(mean_q1, covariance_matrix_q1, 500)

# Question 4: Data (mean = [0, -3])
mean_q4 = [0, -3]
covariance_matrix_q4 = [
    [1.0, 0.8],
    [0.8, 1.0]
]
data_q4 = np.random.multivariate_normal(mean_q4, covariance_matrix_q4, 500)

# Combine the two datasets using np.vstack()
combined_data_vstack = np.vstack((data_q1, data_q4))

# Combine the two datasets using np.concatenate()
combined_data_concatenate = np.concatenate((data_q1, data_q4), axis=0)

# Verify the shape of the combined ndarray
print("Shape using np.vstack():", combined_data_vstack.shape)
print("Shape using np.concatenate():", combined_data_concatenate.shape)

# Check if both methods yield identical results
print("Are the results identical?", np.array_equal(combined_data_vstack, combined_data_concatenate))
```
